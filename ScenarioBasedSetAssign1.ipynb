```python
total = int(input("Enter the total order amount:"))
if(total>100):
    disc = (total*20/100)
    print("you get a discount of $",disc)
    total = total - disc
    print("After discount you pay a total amount of :", total)
else:
    print("no discount")
```

```python
balance = 3000
withdrawal = int(input("Enter the amount to withdraw"))
if(withdrawal <= balance):
    balance = balance-withdrawal
    print("Updated balance is:",balance)
else:
    print("Sorry you cannot withdraw due to insufficient funds")
```

```python
year = int(input("Enter a year"))
if(year%4==0):
    if((year%100 == 0) & (year%400 == 0)):
        print("Its a leap year")
    elif((year%100 > 0) & (year%400 > 0)):        
        print("Its a leap year")
    else:
        print("Its not a leap year")
else:
    print("Its not a leap year")
```
