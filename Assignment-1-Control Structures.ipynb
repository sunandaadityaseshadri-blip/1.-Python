```python
#method 1 - two condition checked with in elif statement
mass = float(input("Enter the BMI Index : ")) 
#if I use "int" instead of "float", the program runs for whole values but not for decimal values
if(mass<=18.4):
    print("Underweight")
elif(mass>=18.5 and mass<=24.9):
    print("normal")
elif(mass>=25 and mass<=39.9):
    print("Overweight")
else:
    print("obese")
```
