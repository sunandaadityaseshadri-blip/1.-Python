```python
#addition
a=43
b=34
print("a =",a)
print("b =",b)
print("Add = ",a+b)
```

```python
#subtraction
a=43
b=34
print("a =",a)
print("b =",b)
print("Sub = ",a-b)
```

```python
#multiplication
a=63
b=97
print("a =",a)
print("b =",b)
print("Mul = ",a*b)
```

```python
#Division - renders along with decimal points
a=76
b=31
print("a =",a)
print("b =",b)
print("Float Div = ",float(a/b)) #both output are same, which means by default the type is float
print("Float Div = ",a/b)
```

```python
#Floor Division - renders only the quotient, leaving back the remainder (basically round in excel)
a=76
b=31
print("a =",a)
print("b =",b)
print("Floar Div = ",a//b)
```

```python
#Modulo - renders the remainder
a=76
b=31
print("a =",a)
print("b =",b)
print("Modulo = ",a%b)
```

```python
#Power
a=4
b=4
print("a =",a)
print("b =",b)
print("Power = ",a**b)
```
