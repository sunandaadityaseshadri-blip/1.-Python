```python
#2.Write a Python program to print the first 5 even numbers. Expected Output:
#2 4 6 8 10
evenNum = []
for num in range(1,11,1):
    if(num%2==0):
        evenNum.append(num)        
print(*evenNum)       
```

```python
#3.Write a Python program to print the sum of numbers from 1 to 10. 
#Expected Output:55

numRange = range(11)
total = 0
for tmp in numRange:
    total +=tmp
print(total)
```

```python
#4.Write a Python function to find the factorial of 5 (5*4*3*2*1=120). Expected Output:120  
facRange = [1,2,3,4,5]
print(facRange)
factorial = 1
for i in facRange:
    factorial *= i
print(factorial)
```
