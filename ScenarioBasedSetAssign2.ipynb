```python
def findFactorial(numFacto):
    if((numFacto == 0) or (numFacto == 1)):
        return 1
    rangeForFacto = range(numFacto,1,-1)
    factOfNum = 1
    for temp in rangeForFacto:
        factOfNum *= temp
    return factOfNum

numFacto = int(input("Enter a number to find its factorial:"))
factorial = findFactorial(numFacto)
print(factorial)
```

```python
def validateArm(inpNum,totalLen):
    #intNum = int(inpNum)
    lst = []
    arm = 0
    for i in inpNum:
        lst.append(i)
    for t in lst:
        arm += int(t)**totalLen
    return arm

inpNum = input("Enter a number to find if its an armstrong number:")
totalLen = len(inpNum)
armstrong = validateArm(inpNum,totalLen)
if(armstrong == int(inpNum)):
    print("Its an armstrong number")
else:
    print("Its not an armstrong number")
```

```python
def swap(stng, stngLen):
    latestStg = []
    for s in stng:
        latestStg.append(s)
    latestStg[0],latestStg[-1] = latestStg[-1], latestStg[0]
    return "".join(latestStg)
        
stng = input("Enter a string :  ")
stngLen = len(stng)
rev = swap(stng, stngLen)
print("first and last char reversed is :", rev)
```

```python
def longestWord(word):
    longest = ""
    for l in word:
        if(len(l) > len(longest)):
            longest = l
    return longest

sent = input("Give me a sentence : ")
word = sent.split()
longest = longestWord(word)
print("The longest word in this sentence is : ", longest)
```
